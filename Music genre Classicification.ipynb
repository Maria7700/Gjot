```python
#Combining audio tracks of the same genre in lists 
disco_data_dir = 'D:\Project Data\genres\disco'
disco_audio_files=glob(disco_data_dir +'/*.wav')

jazz_data_dir='D:\Project Data\genres\jazz'
jazz_audio_files=glob( jazz_data_dir +'/*.wav')

metal_data_dir = 'D:\Project Data\genres\metal'
metal_audio_files=glob(metal_data_dir +'/*.wav')


hiphop_data_dir = 'D:\Project Data\genres\hiphop'
hiphop_audio_files=glob(hiphop_data_dir +'/*.wav')

Rockkk_data_dir = 'D:\Project Data\genres\Rockkk'
Rockkk_audio_files=glob(Rockkk_data_dir +'/*.wav')

pop_data_dir = 'D:\Project Data\genres\pop'
pop_audio_files=glob(pop_data_dir +'/*.wav')

country_data_dir = 'D:\Project Data\genres\country'
country_audio_files=glob(country_data_dir +'/*.wav')

classical_data_dir = 'D:\Project Data\genres\classical'
classical_audio_files=glob(classical_data_dir +'/*.wav')

Reggaeee_data_dir = 'D:\Project Data\genres\Reggaeee'
Reggaeee_audio_files=glob(Reggaeee_data_dir +'/*.wav')

Bluesss_data_dir = 'D:\Project Data\genres\Bluesss'
Bluesss_audio_files=glob(Bluesss_data_dir +'/*.wav')
```

```python
#Combining all audio tracks in a list
z=[]
for i  in range(0,100):
    for j in range(0,10):
        xx=[disco_audio_files[i],jazz_audio_files[i],metal_audio_files[i],hiphop_audio_files[i],Rockkk_audio_files[i],pop_audio_files[i],classical_audio_files[i],Reggaeee_audio_files[i],Bluesss_audio_files[i],country_audio_files[i]]
    
        z.append("".join(xx[j]))
```

```python
z
```

```python
#Defining function that loads and  converts MFCC 2D array of audio track into feature vector
def feat(x,sr):
    x1,sr1 = lr.load(x,sr)
    y=lr.feature.mfcc(x1, sr1,n_mels=12)
    f_list=[]
    index=0
    while index<=11:
        f_list.append(np.mean((y[index])))
        index+=1
    return np.transpose(f_list)
                   
```

```python
#Gathering feature vectors in a list
sr=21966
f=[]
i=0

while i < 1000:
    f.append(feat(z[i],sr))
    i+=1        
    
print(f)
X=np.array(f)


```

```python
#Labeling classes by numbers
genres={'classic':0,'Rock':1, 'jazz':2, 'country':3, 'Blues':4,  'pop':5,  'Reggae':6,  'metal':7, 'disco':8, 'hiphop':9}
```

```python
#Getting label vector
label= [None]*1000
lab=0
while lab<1000:
    if z[lab].find('classic')==23:
        label[lab]=0
    elif z[lab].find('Rock')==23:
        label[lab]=1
    elif z[lab].find('jazz')==23:
        label[lab]=2
    elif z[lab].find('country')==23:
        label[lab]=3
    elif z[lab].find('Blues')==23:
        label[lab]=4
    elif z[lab].find('pop')==23:
        label[lab]=5
    elif z[lab].find('Reggae')==23:
        label[lab]=6
    elif z[lab].find('metal')==23:
        label[lab]=7
    elif z[lab].find('disco')==23:
        label[lab]=8
    elif z[lab].find('hiphop')==23:
        label[lab]=9
        
    lab+=1
Y = np.array(label).reshape(1000,1)

```

```python
X_train,X_test,y_train,y_test=tts(X,Y,test_size=0.2,random_state=24)
```

```python
#Defining function that loads and  converts Chromagram 2D array of audio track into feature vector

def Feat_vector(k,sr1):
    x2,sr2 = lr.load(k,sr1)
    y=lr.feature.chroma_stft(x2, sr2,n_chroma=12)
    f_list1=[]
    index1=0
    while index1<=11:
        f_list1.append(np.mean((y[index1])))
        index1+=1
    return np.transpose(f_list1)
                   
```

```python
#Gathering feature vectors in a list for Chromagram
sr1=21966
f1=[]
d=0

while d< 1000:
    f1.append(Feat_vector(z[d],sr1))
    d+=1
print(f1)
X1=np.array(f1)
```
